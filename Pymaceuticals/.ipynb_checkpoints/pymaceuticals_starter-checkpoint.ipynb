## Observations and Insights

## Dependencies and starter code

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
from sklearn import datasets


# Study data files
mouse_metadata = "data/Mouse_metadata.csv"
study_results = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)

# Combine the data into a single dataset
combined = pd.merge(mouse_metadata, study_results, on='Mouse ID')
combined.head(15)
```

## Summary statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

mean = combined.groupby('Drug Regimen').mean()
median = combined.groupby('Drug Regimen').median()
var = combined.groupby('Drug Regimen').var()
stdev = combined.groupby('Drug Regimen').std()
sem = combined.groupby('Drug Regimen').sem()
mean_df = pd.DataFrame(mean, columns=['Tumor Volume (mm3)'])
mean_df = mean_df.rename(columns={"Tumor Volume (mm3)": "Mean"})
median_df = pd.DataFrame(median, columns=['Tumor Volume (mm3)'])
median_df = median_df.rename(columns={"Tumor Volume (mm3)": "Median"})
mean_med = pd.merge(mean_df, median_df, on='Drug Regimen')
stdev_df = pd.DataFrame(stdev, columns=['Tumor Volume (mm3)'])
stdev_df = stdev_df.rename(columns={"Tumor Volume (mm3)": "Standard Deviation"})
mmst = pd.merge(mean_med, stdev_df, on='Drug Regimen')
var_df = pd.DataFrame(var, columns=['Tumor Volume (mm3)'])
var_df = var_df.rename(columns={"Tumor Volume (mm3)": "Variance"})
mmstv = pd.merge(mmst, var_df, on='Drug Regimen')
sem_df = pd.DataFrame(sem, columns=['Tumor Volume (mm3)'])
sem_df = sem_df.rename(columns={"Tumor Volume (mm3)": "SEM"})
drug_df = pd.merge(mmstv, sem_df, on='Drug Regimen')
drug_df.head(10)
```

## Quartiles, outliers and boxplots

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens. Calculate the IQR and quantitatively determine if there are any potential outliers. 
```

```python
grouped = combined.groupby(['Mouse ID']).max()
grouped = grouped.reset_index()
grouped.head()
#merge Mouse ID and timepoint to original df (left merge)
grouped = grouped[['Mouse ID', 'Timepoint']].merge(combined, on=['Mouse ID', 'Timepoint'], how='left')
grouped.head()
#Capomulin, Ramicane, Infubinol, and Ceftamin
ramicane_final = grouped.loc[grouped['Drug Regimen']=='Ramicane']['Tumor Volume (mm3)']
cap_final = grouped.loc[grouped['Drug Regimen']=='Capomulin']['Tumor Volume (mm3)']
inf_final = grouped.loc[grouped['Drug Regimen']=='Infubinol']['Tumor Volume (mm3)']
ceft_final = grouped.loc[grouped['Drug Regimen']=='Ceftamin']['Tumor Volume (mm3)']
ram_df = pd.DataFrame(ramicane_final)
final_volume_df = grouped[['Drug Regimen', 'Tumor Volume (mm3)']]
#list of final tumor volumes - mice on Capomulin, Ramicane, Infubinol, and Ceftamin regimens
final_volume_df.head(15)
```

```python
#IQRs
ramicane_quartiles = ramicane_final.quantile([.25,.5,.75])
ramicane_lower = ramicane_quartiles[.25]
ramicane_upperq = ramicane_quartiles[0.75]
ramicane_iqr = ramicane_upperq-ramicane_lower
ramicane_lower_bound = ramicane_lower - (1.5*ramicane_iqr)
ramicane_upper_bound = ramicane_upperq + (1.5*ramicane_iqr)

cap_quartiles = cap_final.quantile([.25,.5,.75])
cap_lower = cap_quartiles[.25]
cap_upperq = cap_quartiles[0.75]
cap_iqr = cap_upperq-cap_lower
cap_lower_bound = cap_lower - (1.5*cap_iqr)
cap_upper_bound = cap_upperq + (1.5*cap_iqr)

inf_quartiles = inf_final.quantile([.25,.5,.75])
inf_lower = inf_quartiles[.25]
inf_upperq = inf_quartiles[0.75]
inf_iqr = inf_upperq-inf_lower
inf_lower_bound = inf_lower - (1.5*inf_iqr)
inf_upper_bound = inf_upperq + (1.5*inf_iqr)

ceft_quartiles = ceft_final.quantile([.25,.5,.75])
ceft_lower = ceft_quartiles[.25]
ceft_upperq = ceft_quartiles[0.75]
ceft_iqr = ceft_upperq-ceft_lower
ceft_lower_bound = ceft_lower - (1.5*ceft_iqr)
ceft_upper_bound = ceft_upperq + (1.5*ceft_iqr)
print(f"IQR of Capomulin is {cap_iqr}, the IQR of Ramicane in {ramicane_iqr}, the IQR of Infubinol is {inf_iqr}, and the IQR of Ceftamin is {ceft_iqr}")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
color=dict(markerfacecolor='green', markersize=14)
plt.boxplot([ramicane_final, cap_final, inf_final, ceft_final],  labels=['Ramicane', 'Capomulin', 'Infubinol', 'Ceftamin'],  flierprops=color)
plt.show
```

## Line and scatter plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin

#cap_final = grouped.loc[grouped['Drug Regimen']=='Capomulin']['Tumor Volume (mm3)']

capo185 = combined.set_index('Drug Regimen')
capo185 = capo185.loc['Capomulin',:]
capo185 = capo185.reset_index()
capo185 = capo185.loc[capo185['Mouse ID']=='s185']
time = capo185['Timepoint']
tumor = capo185['Tumor Volume (mm3)']
# gender_chart.set_xlabel("Gender")
# gender_chart.set_ylabel("Number of Trips Taken")
capo_line = plt.plot(time, tumor, color="blue", label="Mouse s185 Tumor Size")
plt.xlabel("Timpoint")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Time Point vs. Tumor Volume - Mouse s185")
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
capo_group = combined.set_index('Drug Regimen')
capo_group = capo_group.reset_index()
capo_group = capo_group.loc[capo_group['Drug Regimen']=='Capomulin']

capo_ave = capo_group.groupby(capo_group['Mouse ID']).mean()
capo_ave = capo_ave.reset_index()
plt.scatter(capo_ave['Weight (g)'], capo_ave['Tumor Volume (mm3)'])
plt.xlabel("Weight of Mouse in Grams")
plt.ylabel("Volume of Tumor (mm3)")
plt.title("Mouse Weight vs. Average Tumor Volume")
```

```python
capo_ave.head(20)
```

```python
# Calculate the correlation coefficient and linear regression model for mouse weight and average tumor volume for the Capomulin regimen
x = capo_ave.iloc[:,2]
y = capo_ave.iloc[:,4]
correlation = st.pearsonr(x, y)
print(f"The correlation between both factors is {round(correlation[0],2)}")


(slope, intercept, rvalue, pvalue, stderr) = linregress(x, y)
regress_values = x * slope + intercept
```

```python

plt.scatter(x, y)
plt.plot(x, regress_values, "--")

plt.xlabel("Weight")
plt.ylabel("Tumor Volume")
plt.show()
```
